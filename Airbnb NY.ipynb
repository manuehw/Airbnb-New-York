```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("AB_NYC_2019.csv")
```

```python
df.head()
```

```python
# Vemos la media de precio por noche según las distintas zonas de NY
#Manhattan es la mas cara y el Bronk es la zona mas barata
df.groupby("neighbourhood_group")[["price"]].mean()
```

```python
#Obtenemos la media de los precios por barrio en NY
tab=pd.pivot_table(df,index=df["neighbourhood"], aggfunc={"price":"mean"} ).fillna(0)
tab.tail(16)
```

```python
#Obtenemos la media de días disponibles en NY según las zonas  
tab=pd.pivot_table(df,index=df["neighbourhood_group"], aggfunc={"availability_365":"mean"} ).fillna(0)
tab
```

```python
#Calculamos la media del mínimo de noches en airbnb
tab=pd.pivot_table(df,index=df["neighbourhood_group"], aggfunc={"minimum_nights":"mean"} ).fillna(0)
tab
```

```python
#media de precios por tipo de habitación
tab2=pd.pivot_table(df,index=df["room_type"], aggfunc={"price":"mean"} ).fillna(0)
tab2
```

```python
#media de precios por tipo de habitación y zona de NY
tab2=pd.pivot_table(df,index=[df["room_type"],df["neighbourhood_group"]], aggfunc={"price":"mean"} ).fillna(0)
tab2
```

```python
#El precio más comun por tipo de habitación y zona de NY
tab3=pd.pivot_table(df,index=[df["room_type"],df["neighbourhood_group"]], aggfunc={"price":"median"} ).fillna(0)
tab3
```
